```python
#creating  dataframes to find win rate
##creating hwin_rate dataframe 

win_rate= games_unique[['away_team_id','home_team_id','outcome']]
win_rate
```

```python
unique_outcome= win_rate['outcome'].unique()
unique_outcome
```

```python
win_rate= win_rate.sort_values(by='away_team_id')
win_rate.tail(50)
```

```python

win_rate[['winner','REG/OT']]= win_rate['outcome'].str.split(n=1,expand=True)

win_rate.drop(columns=['outcome'], inplace=True)

win_rate
    
#split the column outcome to two columns to do the analysis in an easy way.
```

```python
tbc_rows= win_rate[win_rate['winner']=='tbc']
tbc_rows
#checked the data frames for rows with value tbc which is either draw or winner is not decided
```

```python

# Create a list of all team IDs
all_teams = pd.concat([win_rate['away_team_id'], win_rate['home_team_id']])

# Count the occurrences of each team ID
#
#all_teams
games_played =all_teams.value_counts()

games_played = pd.DataFrame({'team_id':games_played.index,'total_games_played':games_played.values})
games_played
```

```python
games_played= games_played.sort_values(by='team_id')
games_played
```

```python
home_games = win_rate.groupby('home_team_id').size().rename('home_games')
home_games_df= pd.DataFrame({'team_id':home_games.index,'home_games_played':home_games.values})

away_games = win_rate.groupby('away_team_id').size().rename('away_games')
away_games_df= pd.DataFrame({'team_id':away_games.index,'away_games_played':away_games.values})

# Count the number of wins for each team at home and away
home_wins = win_rate[win_rate['winner'] == 'home'].groupby('home_team_id').size().rename('home_wins')
home_wins_df= pd.DataFrame({'team_id':home_wins.index,'home_games_won':home_wins.values})

away_wins = win_rate[win_rate['winner'] == 'away'].groupby('away_team_id').size().rename('away_wins')
away_wins_df= pd.DataFrame({'team_id':away_wins.index,'away_games_won':away_wins.values})

# Calculate win rates for each team at home and away
home_win_rate = (home_wins / home_games) * 100
away_win_rate = (away_wins / away_games) * 100

# Combine the results into a single DataFrame
win_rate_summary = pd.DataFrame({'home_win_rate': home_win_rate, 'away_win_rate': away_win_rate})

win_rate_summary.reset_index(inplace=True)
win_rate_summary.rename(columns={'index':'team_id'}, inplace=True)
win_rate_summary
```

```python
home_play =pd.merge(home_games_df,home_wins_df, on ='team_id')
away_play= pd.merge(away_games_df,away_wins_df, on ='team_id')
home_away_games= pd.merge(home_play,away_play,on ='team_id')
home_away_games_rate= pd.merge(home_away_games,win_rate_summary, on='team_id')
home_away_games_rate
#merged different dataframes to form a single data frame.
```

```python
home_away_games_rate.drop(home_away_games_rate.index[-1], inplace=True)
```

```python

plt.figure(figsize=(10,6))
plt.plot(home_away_games_rate['team_id'],home_away_games_rate['home_win_rate'],marker='o',linestyle='-', label='Home win rate')
plt.plot(home_away_games_rate['team_id'],home_away_games_rate['away_win_rate'],marker='o',linestyle='-',label='Away win rate')

plt.xlabel('Team ID')
plt.ylabel('Win rate%')
plt.title('Home Vs Away win Rate for Each Team')

plt.legend()
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

```python
# Plotting
plt.figure(figsize=(10, 6))

# Set the width of the bars
bar_width = 0.35

# Define the positions of the bars on the x-axis
x = range(len(home_away_games_rate))

# Plot home win rate for each team as bars
plt.bar(x, home_away_games_rate['home_win_rate'], width=bar_width,color='0.2', label='Home Win Rate')

# Plot away win rate for each team as bars
plt.bar([i + bar_width for i in x], home_away_games_rate['away_win_rate'], width=bar_width,color='blue', label='Away Win Rate')

# Add labels and title
plt.xlabel('Team ID')
plt.ylabel('Win Rate (%)')
plt.title('Home vs. Away Win Rate for Each Team')
plt.xticks([i + bar_width / 2 for i in x], home_away_games_rate['team_id'], rotation=90)  # Set x-axis tick labels
plt.legend()

# Show plot
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```
