```python
#Iterate over each column in the DataFrame
for column in games.columns:
    # Get the unique values for the current column
    unique_values = games[column].unique()
    # Print the column name and its unique values
    print(f"Unique values for column '{column}':")
    print(unique_values)
    print()  # Add a newline for readability
```

```python
 #Check for duplicate rows in the DataFrame
duplicate_rows = games.duplicated()

# Sum the number of duplicate rows
num_duplicates = duplicate_rows.sum()

# Check if there are any duplicate rows
if num_duplicates == 0:
    print("The DataFrame is unique.")
else:
    print("The DataFrame contains duplicate rows.")
```

```python
duplicate=games[duplicate_rows]
print("Duplicate rows:")
duplicate
```

```python
# Remove duplicate rows from the DataFrame
games_unique = games.drop_duplicates()

# Print the DataFrame without duplicate rows
print("DataFrame without duplicate rows:")
games_unique
```

```python
data_types=games_unique.dtypes
data_types
```

```python

# Convert 'date_time_GMT' column to datetime format
games_unique['date_time_GMT'] = pd.to_datetime(games_unique['date_time_GMT'], format='%Y-%m-%dT%H:%M:%SZ')

# Convert datetime format to 'mm-dd-yyyy hh:mm' format
games_unique['date_time_GMT'] = games_unique['date_time_GMT'].dt.strftime('%m-%d-%Y %H:%M')

# Display the DataFrame with the updated 'date_time_GMT' column
#print(games)
games_unique.head(50)
```

```python
home_match=games_unique[['home_team_id','home_goals']]
home_match
```

```python
home_match= home_match.sort_values(by='home_team_id')
home_match.head(50)

```

```python
home_match
```

```python
avg_home_goals= home_match.groupby('home_team_id')['home_goals'].mean()
avg_home_goals_df=avg_home_goals.reset_index()
avg_home_goals_df.columns=['home_team_id','average_home_goals']
avg_home_goals_df
```

```python
dt= avg_home_goals_df.dtypes
dt
```

```python

avg_home_goals_df.to_csv('C:/Users/Ros/Desktop/JDC/mini project/averge_home_goals.csv')
avg_home_goals_df.to_excel('C:/Users/Ros/Desktop/JDC/mini project/averge_home_goals_df.xlsx')
```

```python
import matplotlib.pyplot as plt
avg_home_goals_df['home_team_id']= avg_home_goals_df['home_team_id'].astype(str)
plt.figure(figsize=(10,6))
plt.bar(avg_home_goals_df['home_team_id'],avg_home_goals_df['average_home_goals'])
plt.title('Average Home Goals for Each Home Team')
plt.xlabel('Home Team ID')
plt.ylabel('Average home goals')
plt.xticks(rotation=90)
plt.show()
```

```python
away_match=games_unique[['away_team_id','away_goals']]
away_match
```

```python

away_match= away_match.sort_values(by='away_team_id')
away_match.head(50)
```

```python
avg_away_goals= away_match.groupby('away_team_id')['away_goals'].mean()
avg_away_goals_df=avg_away_goals.reset_index()
avg_away_goals_df.columns=['away_team_id','average_away_goals']
avg_away_goals_df
```
